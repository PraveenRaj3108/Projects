#### 3. Display the percentage distribution of the data in each year using the pie chart.

```python
year_distribution = df['year'].value_counts(normalize=True) * 100

plt.figure(figsize=(8, 8))
plt.pie(year_distribution, labels=year_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage Distribution of Data by Year')
plt.show()
```

#### 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings

```python

plt.figure(figsize=(10, 6))
sns.histplot(df['priceperlb'], kde=True, bins=20)
plt.title('Distribution of Price')
plt.xlabel('Price Per lb')
plt.ylabel('Frequency')
plt.show()
```

```python
## This shows histogram that represents the distribution of the "price per lb" variable.And it shows the centraltendency
##,And how the data is skewed and also shows the distribution of the data
```

#### 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='numcol', y='prodvalue', data=df)
plt.title('Relationship between numcol and prodvalue')
plt.xlabel('Number of Col')
plt.ylabel('Prodvalue')
plt.show()
```

```python
#By seeing the Scatter plot we are able to see that the relationship between the numcol and prodvalue are increase in the trend
#and till 100000 there is strong relationship
```

#### 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='year', y='prodvalue', data=df)
plt.title('Relationship between year and prodvalue')
plt.xlabel('Year')
plt.ylabel('Prodvalue')
plt.show()
```

```python
# By seeing this we can see in each year there are more outliers in the relationship between year and prodvalue
```

#### 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
columns = ['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year']

sns.pairplot(df[columns], hue='year', palette='viridis')
plt.suptitle(' Pairplot of Year', y=1.02)
plt.show()
```

```python
#This shows how the multiple pairs of variables are correlated with each of them which some variables have strong
#correlation and with increase in trend
```

#### 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol','total prod', 'stocks', 'price per lb', 'prodvalue')


```python
columns1 = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']

correlation = df[columns1].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
